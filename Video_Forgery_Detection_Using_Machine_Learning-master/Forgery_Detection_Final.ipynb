```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import tensorflow as tf
```

```python
tf.config.run_functions_eagerly(True)
```

```python
vid_src = r"C:\Users\Fahmi Bakhir\Desktop\Video_Forgery_Detection_Using_Machine_Learning-master\Video_Forgery_Detection_Using_Machine_Learning-master\video_tampering_dataset\videos\Forged\01_forged.mp4"
vid = []

sumframes =0
cap= cv2.VideoCapture(vid_src)
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    sumframes +=1
    vid.append(frame)
cap.release()
    
print("No. Of Frames in the Video: ",sumframes)
```

```python
Xtest = np.array(vid)
#print(Xtest.shape)
```

```python
from keras.models import load_model
```

```python
model = load_model(r"C:\Users\Fahmi Bakhir\Desktop\Video_Forgery_Detection_Using_Machine_Learning-master\Video_Forgery_Detection_Using_Machine_Learning-master/ResNet50_Model/forgery_model.hdf5")
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', run_eagerly=True)
```

```python
output = model.predict(Xtest)
```

```python
print(output.shape)
```

```python
output = output.reshape((-1))
#print(output.shape)
```

```python
results = []
for i in output:
    if i>0.5:
        results.append(1)
    else:
        results.append(0)
```

```python
print(len(results))
print(results)
```

```python
forge_flag = 0
for i in results:
    if i == 1:
        forge_flag = 1
        break
        
if forge_flag == 0:
    print("The video is not forged")
    
else:
    print("The video is forged")
    print("Number of Forged Frames in the video: ",sum(results))
```
